```python
import pandas as pd

# Define una funcin para convertir los valores en millones y billones a nmeros reales
def convert_to_float(value):
    if 'M' in value:
        return float(value.replace('$', '').replace('M', '')) * 1e6
    elif 'B' in value:
        return float(value.replace('$', '').replace('B', '')) * 1e9
    else:
        return float(value.replace('$', ''))

# Verifica el resultado
print(df.dtypes)
```

```python
df['Total Raised'] = df['Total Raised'].apply(convert_to_float)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado df

# Filtra solo las columnas numricas
df_numeric = df.select_dtypes(include='number')

# Crea el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```
